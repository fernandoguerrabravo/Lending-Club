```python
#Verify the Nan Values in the columns
df.isnull().sum().sort_values(ascending=False)
```

```python
#Function to Check Nan in % 
def missing_check(df):
    total = df.isnull().sum().sort_values(ascending=False)   # total number of null values
    percent = ((df.isnull().sum()/df.isnull().count())*100).sort_values(ascending=False)  # percentage of values that are null
    types = df.dtypes
    missing_data = pd.concat([total, percent, types], axis=1, keys=['Total', 'Percent', 'Types'])  # putting the above two together
    return missing_data # return the dataframe
missing_check(df)
```

```python
#Delete Columns over 30% Missing Values
columns_missing_30 = missing_check(df)[missing_check(df)['Percent'] > 30]
columns_missing_30
```

```python
#List Columns Missing Values > 30%
index = columns_missing_30.index
index
```

```python
#Drop Columns with Missing Values > 30%
df = df.drop(index, axis=1)
#Data Frame Shape after Drop
print(df.shape)
```

```python
#Check Columns with Missing Values > 0%
columns_missing_0 = missing_check(df)[missing_check(df)['Percent'] > 0]
columns_missing_0
```

```python
#Clean Data Frame to Analisys
df.head()
```

```python
#Features Transform Type 
#term: Object to numeric
#Filter and Drop Missing Values
df = df[~df['term'].isnull()]
# Extract numeric values from the string
import re
df['term'] = df['term'].apply(lambda x: re.findall('\d+', str(x))[0])
# convert to numeric
df['term'] = df['term'].apply(lambda x: pd.to_numeric(x))
df['term'].dtype
```

```python
#emp_length: Object to numeric
#Filter and Drop Missing Values
df = df[~df['emp_length'].isnull()]
# Extract numeric values from the string
import re
df['emp_length'] = df['emp_length'].apply(lambda x: re.findall('\d+', str(x))[0])
# convert to numeric
df['emp_length'] = df['emp_length'].apply(lambda x: pd.to_numeric(x))
df['emp_length'].dtype
```

```python
# The column int_rate is STR, let's convert it to float
df['int_rate'] = df['int_rate'].apply(lambda x: pd.to_numeric(x.split("%")[0]))
```

```python
# Loan Status  Transform to Category 
df['loan_status'] = df['loan_status'].astype('category')
df['loan_status'].value_counts()
```

```python
# filter charged-off,  we not use Current Status
df = df[df['loan_status'] != 'Current']
df['loan_status'] = df['loan_status'].apply(lambda x: 0 if x=='Fully Paid' else 1)

# converting loan_status to integer type
df['loan_status'] = df['loan_status'].apply(lambda x: pd.to_numeric(x))

# summarising the values
df['loan_status'].value_counts()
```
